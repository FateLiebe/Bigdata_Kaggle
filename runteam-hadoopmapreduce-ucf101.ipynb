```python
%%bash
# ==============================================================================
# GIAI ĐOẠN 0: THIẾT LẬP SCRIPT MAPREDUCE (PHIÊN BẢN CUỐI CÙNG)
# ==============================================================================
echo "--- GIAI ĐOẠN 0: THIẾT LẬP SCRIPT MAPREDUCE ---"
cd /kaggle/working/TEAM

# --- File 1: mapper.py (Tạo cấu trúc 3 cấp và trích xuất khung hình) ---
cat > mapper.py << 'EOF'
#!/usr/bin/env python3.10
import sys, os, subprocess

# ĐƯỜNG DẪN VIDEO GỐC
VIDEO_SOURCE_DIR = "/kaggle/working/TEAM/data/hmdb51_videos"
# ĐÂY LÀ THƯ MỤC OUTPUT CUỐI CÙNG MÀ LỆNH TRAIN CẦN
FINAL_OUTPUT_DIR = "/kaggle/working/hmdb51_frames"
# FILE GHI LOG LỖI
FFMPEG_LOG_FILE = "/kaggle/working/ffmpeg_errors.log"

def main():
    # Xác định loại split từ tên file input của Hadoop
    filepath = os.environ.get("map_input_file", "")
    split_type = "unknown"
    if "train" in filepath: split_type = "train"
    elif "val" in filepath: split_type = "val"
    elif "test" in filepath: split_type = "test"

    for line in sys.stdin:
        # Tương thích với định dạng "class/video_name id"
        line = line.strip().split(" ")[0]
        if not line or '/' not in line: continue
        try:
            video_name_no_ext = os.path.splitext(os.path.basename(line))[0]
            class_name = os.path.basename(os.path.dirname(line))
            
            source_video_path = os.path.join(VIDEO_SOURCE_DIR, class_name, f"{video_name_no_ext}.avi")
            if not os.path.exists(source_video_path): continue

            # TẠO CẤU TRÚC THƯ MỤC 3 CẤP MÀ video_reader.py YÊU CẦU
            # Ví dụ: /kaggle/working/hmdb51_frames/train/brush_hair/v_brush_hair_g01_c01
            target_frame_dir = os.path.join(FINAL_OUTPUT_DIR, split_type, class_name, video_name_no_ext)
            os.makedirs(target_frame_dir, exist_ok=True)

            # Trích xuất khung hình vào đúng thư mục đích
            ffmpeg_cmd = ['ffmpeg', '-y', '-i', source_video_path, '-threads', '1', '-q:v', '0', f'{target_frame_dir}/%06d.jpg']
            result = subprocess.run(ffmpeg_cmd, capture_output=True, text=True, timeout=90)

            if result.returncode == 0:
                # Output để Reducer biết video nào đã xử lý xong
                print(f"{split_type}\t{class_name}/{video_name_no_ext}")
            else:
                 with open(FFMPEG_LOG_FILE, "a") as log:
                    log.write(f"--- FFMPEG FAILED on {source_video_path} ---\nSTDERR: {result.stderr}\n")
        except Exception as e:
            with open(FFMPEG_LOG_FILE, "a") as log:
                log.write(f"--- PYTHON ERROR on line {line}: {e} ---\n")

if __name__ == "__main__":
    main()
EOF
echo "Đã tạo file mapper.py"

# --- File 2: reducer.py (Chỉ tổng hợp báo cáo) ---
cat > reducer.py << 'EOF'
#!/usr/bin/env python3.10
import sys

def main():
    train_count = 0
    val_count = 0
    test_count = 0

    for line in sys.stdin:
        try:
            split_type, video_path = line.strip().split('\t')
            if split_type == "train":
                train_count += 1
            elif split_type == "val":
                val_count += 1
            elif split_type == "test":
                test_count += 1
        except ValueError:
            continue

    # In báo cáo ra stdout (sẽ được lưu vào HDFS)
    print("--- Job Summary ---")
    print(f"Successfully processed 'train' videos: {train_count}")
    print(f"Successfully processed 'val' videos: {val_count}")
    print(f"Successfully processed 'test' videos: {test_count}")

if __name__ == "__main__":
    main()
EOF
echo "Đã tạo file reducer.py"

# ==============================================================================
# GIAI ĐOẠN 1: CHẠY JOB MAPREDUCE DUY NHẤT
# ==============================================================================
echo "--- GIAI ĐOẠN 1: CHUẨN BỊ VÀ CHẠY JOB MAPREDUCE TOÀN BỘ ---"

# Dọn dẹp môi trường, tạo thư mục output cuối cùng
rm -rf /kaggle/working/hmdb51_frames
mkdir -p /kaggle/working/hmdb51_frames
rm -f /kaggle/working/ffmpeg_errors.log
hdfs dfs -rm -r -skipTrash /user/root/team_project_input || true
hdfs dfs -rm -r -skipTrash /user/root/team_project_output || true

# Chuẩn bị input cho Hadoop
hdfs dfs -mkdir -p /user/root/team_project_input
hdfs dfs -put splits/hmdb/trainlist.txt /user/root/team_project_input/mapreduce_train.txt
hdfs dfs -put splits/hmdb/vallist.txt /user/root/team_project_input/mapreduce_val.txt
hdfs dfs -put splits/hmdb/testlist.txt /user/root/team_project_input/mapreduce_test.txt

# Thiết lập biến môi trường Hadoop
export HADOOP_HOME=/kaggle/working/hadoop; export PATH=$HADOOP_HOME/bin:$HADOOP_HOME/sbin:$PATH
export JAVA_HOME=/usr/lib/jvm/java-11-openjdk-amd64
# ... (các biến môi trường khác)

# Chạy job MapReduce
echo "Bắt đầu job MapReduce... CẢNH BÁO: RỦI RO CAO BỊ LỖI ĐẦY Ổ ĐĨA."
hadoop jar $HADOOP_HOME/share/hadoop/tools/lib/hadoop-streaming-*.jar \
    -D mapred.job.name="Full_Preprocessing_Final" \
    -D mapreduce.job.reduces=1 \
    -files mapper.py,reducer.py \
    -input /user/root/team_project_input/ \
    -output /user/root/team_project_output \
    -mapper "python3.10 mapper.py" \
    -reducer "python3.10 reducer.py"

if [ $? -ne 0 ]; then
    echo "LỖI: Job MapReduce thất bại! Rất có thể là do hết dung lượng ổ đĩa."
    echo "Kiểm tra file /kaggle/working/ffmpeg_errors.log để biết thêm chi tiết."
    exit 1
fi

# ==============================================================================
# GIAI ĐOẠN 2: HOÀN TẤT
# ==============================================================================
echo ""
echo "--- TIỀN XỬ LÝ HOÀN TẤT ---"
echo "Dữ liệu khung hình đã được tạo tại: /kaggle/working/hmdb51_frames"
```
